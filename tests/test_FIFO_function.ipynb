```python
import numpy as np
from scipy.signal import argrelextrema
import time

import matplotlib.pyplot as plt
plt.style.use('ggplot')
% matplotlib inline
```

```python
N = 100
N1 = np.full(100, False)
N2 = np.full(100, False)
N2[78:85] = True
N1[49:54] = True
N2[49:54] = True

x = np.arange(0,N)
W = 10*np.sin(x/5) + x
W[78:85] = np.inf
W[49:54] = np.inf
```

```python
def FIFO_maker(y, N1):
    y = 1.0 * W
    arg = np.squeeze(argrelextrema(y, np.less))
    if arg.shape == ():
        arg = np.array([arg])
    else:
        None
    y_FIFO = 1*y
    for a in arg:
        loc = np.argwhere(y[:a+1] <= y[a])[-2:] 
        if loc.shape == (2,1):
            if True in N1[int(loc[0]):int(loc[1])]:
                None
            else:
                y_FIFO[int(loc[0]):int(loc[1])] = y[a]
    return y_FIFO
```

```python
plt.plot(W)
plt.plot(FIFO_maker(W,N1))
```
